**Scratch Code MAE,MSE,RMSE**







```python
import numpy as np
```

```python
ac = np.array([10, 20, 30, 40, 50])
pr = np.array([11, 18, 33, 45, 47])
```

```python
print(ac)
```

```python
print(pr)
```

```python
dif = ac - pr
```

```python
print(dif)
```

**MAE**

```python
x = abs(dif)
print(x)
```

```python
mae = np.mean(x)
print(mae)
```

**MSE**

```python
s = np.square(dif)
```

```python
print(s)
```

```python
mse = np.mean(s)
print(mse)
```

**RMSE**

```python
rmse = np.square(mse)
print(rmse)
```

**Using Library**

**MAE**

```python
from sklearn.metrics import mean_absolute_error
```

```python
y_true = [10, 20, 30, 40, 50]
y_pred = [11, 18, 33, 45, 47]
```

```python
MAE = mean_absolute_error(y_true, y_pred)
print(MAE)
```

**MSE**

```python
from sklearn.metrics import mean_squared_error
```

```python
mse = mean_squared_error(y_true, y_pred)
print(mse)
```

**RMSE**

```python
from sklearn.metrics import mean_squared_error
```

```python
rmse = mean_squared_error(y_true, y_pred, squared = False)
print(rmse)
```

**CROSS ENTROPY LOG LOSS**

```python
def sigmoid(z):
    return 1.0 / (1.0 + np.exp(-z))
 
'''
yHat represents the predicted value / probability value calculated as output of hypothesis / sigmoid function
 
y represents the actual label
'''
def cross_entropy_loss(yHat, y):
    if y == 1:
      return -np.log(yHat)
    else:
      return -np.log(1 - yHat)
```

```python
# Calculate sample values for Z
z = np.arange(-10, 10, 0.1)
# Calculate the hypothesis value / probability value
h_z = sigmoid(z)
# Value of cost function when y = 1
# -log(h(x))
cost_1 = cross_entropy_loss(h_z, 1)
# Value of cost function when y = 0
# -log(1 - h(x))
cost_0 = cross_entropy_loss(h_z, 0)
```

```python
print(cost_1)
```

```python
print(cost_0)
```
